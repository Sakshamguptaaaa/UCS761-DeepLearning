```python
import pandas as pd
import numpy as np
# Load dataset
data = pd.read_csv("multiple_linear_regression_dataset.csv")

# Generate a synthetic 'TestScore' column with random integers between 60 and 100
data['TestScore'] = np.random.randint(60, 101, size=len(data))

# Inspect data
print(data.head())
print(data.columns)
print(data.shape)
```

```python
# Inputs (features)
X = data[["experience","TestScore"]]
# Output (target)
y = data["income"]
```

```python
n_features = X.shape[1]
# Initialize weights and bias
w = np.zeros(n_features)
b =0.0
     
```

```python
"""
 X : input features
 w : weights
 b : bias
 returns predicted values

"""
def predict(X, w, b):
 y_hat = X.dot(w) + b
 return y_hat
```

```python
"""
 y : actual values
 y_hat : predicted values

"""
def mean_squared_error(y, y_hat):
  loss = ((y_hat - y) **2).mean()
  return loss
```

```python

def compute_gradients(X, y, y_hat):

  N =len(y)
  # Gradient w.r.t weights
  dw = (2 / N) * X.T.dot(y_hat - y)
  # Gradient w.r.t bias
  db = (2 / N) * (y_hat - y).sum()
  return dw, db

     
```

```python
def update_parameters(w, b, dw, db, lr):
  w = w - lr * dw
  b = b - lr * db
  return w, b
```

```python
lr =0.0001
epochs =1000
```

```python
for epoch in range(epochs):
 y_hat = predict(X, w, b)
 loss = mean_squared_error(y, y_hat)
 dw, db = compute_gradients(X, y, y_hat)
 w, b = update_parameters(w, b, dw, db, lr)

 if epoch %100 ==0:
  print(f"Epoch {epoch}, Loss:{loss}")
```

```python
new_candidate = np.array([4.5,68])
predicted_salary = new_candidate.dot(w) + b
print(predicted_salary)
```
