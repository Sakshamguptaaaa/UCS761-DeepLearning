```python
# definig activation fn to get probability

import numpy as np

def sigmoid(x):
    return 1 / 1 + np.exp(-x)
```

```python
# predicting prob of a record to belong to class 1

def predict_prob(x, w, b):
    z = x@w + b
    p = sigmoid(z)
    return p
```

```python
# error from real value

def loss():
    return -np.mean(y*np.log(p) + (1-y)*np.log(1-p))
```

```python
#update weights for better result

def update_weights(x, w, b, lr, y):
    p = predict_prob(x, w, b)
    error = p - y

    w = w - lr*(x.T @ error) / len(y)
    b = b - lr*np.mean(error)

    return w, b
```

```python
w = np.zeros(x_train.shape[1])
lr = 0.1
b = 0
epochs = 100

for _ in range(epochs):
    w, b = update_weights(x, w, b, lr, y)
```
